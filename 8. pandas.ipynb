## Groupings

```python
quality_groupings = wine_data.groupby('quality').agg({'alcohol':'mean'})
quality_groupings
```

```python
quality_groupings = wine_data.groupby('quality', as_index=False).agg({'alcohol':'mean'})
quality_groupings
```

```python
def rating_function(x):
    if x >= 5.0:
        return "good"
    else:
        return "fair"
```

```python
wine_data["quality_category"] = wine_data["quality"].apply(rating_function)

wine_data.head()
```

```python
wine_data["quality_category 2"] = wine_data["quality"].apply(lambda x: 'good' if x > 6 else 'bad')

wine_data.head(2)
```
